## Задание 1

```python
def doaction(word):  
    for i in range(len(word)-2):
        print(word[i:i+2])
    print(word[-2:])   
```

```python
sentence = ["абракадабра", "фокус", "покус", "десятка", "за", "курс", "немедленно", "появись"]
spisok = []
for word in sentence:
    for i in range(len(word)-2):
        spisok.append(word[i:i+2])
    spisok.append(word[-2:])
    print(spisok)
```

```python
p = {}
for c in spisok:
    if c in p:
        p[c] = p[c]+1
    else:
        p[c]= 1
print (p)
        
        
```

```python
for key in p:  
    if key[0] == key[1]: 
        print("Биграмма из двух одинаковых букв!")
        break
    else:
        print(key, p[key])
```

Даны два слова, ваша программа должна проверять является ли второе слово обратной записью первого слова.

```python
def function():
    word_1 = input()
    word_2 = input()
    if word_1 == word_2[::-1]:
        print("yes")
    else:
        print("no")
function()
function()
        
```
